```python
import pandas as pd
import numpy as np
```

```python
# create a sample DataFrame
data = {'col1': [1, 2, 3, 4, 5], 'col2': [6, 7, 8, 9, 10], 'col3': [11, 12, 13, 14, 15]}
df = pd.DataFrame(data)
```

```python
# apply NumPy functions to DataFrame columns
df['mean_col'] = df['col1'].apply(np.mean)  # calculate mean of column
df['median_col'] = df['col2'].apply(np.median)  # calculate median of column
df['std_col'] = df['col3'].apply(np.std)  # calculate standard deviation of column
df['min_col'] = df['col1'].apply(np.min)  # calculate minimum of column
df['max_col'] = df['col2'].apply(np.max)  # calculate maximum of column
df['var_col'] = df['col3'].apply(np.var)  # calculate variance of column
df['sin_col'] = df['col1'].apply(np.sin)  # apply sine function to column
df['cos_col'] = df['col2'].apply(np.cos)  # apply cosine function to column
df['exp_col'] = df['col3'].apply(np.exp)  # apply exponential function to column
df['log_col'] = df['col1'].apply(np.log)  # apply natural logarithm function to column
df['abs_col'] = df['col2'].apply(np.abs)  # apply absolute value function to column
df['square_col'] = df['col3'].apply(np.square)  # apply square function to column
df['sqrt_col'] = df['col1'].apply(np.sqrt)  # apply square root function to column
df['power_col'] = df['col2'].apply(lambda x: np.power(x, 2))  # apply power function to column
df['round_col'] = df['col3'].apply(lambda x: np.round(x, 2))  # round column to 2 decimal places
df['log1p_col'] = df['col1'].apply(np.log1p)  # apply natural logarithm of (1+x) function to column
df['log10_col'] = df['col2'].apply(np.log10)  # apply base-10 logarithm function to column
df['expm1_col'] = df['col3'].apply(np.expm1)  # apply exponential minus one function to column
df['sign_col'] = df['col1'].apply(np.sign)  # apply sign function to column
df['ceil_col'] = df['col2'].apply(np.ceil)  # apply ceiling function to column
df['floor_col'] = df['col3'].apply(np.floor)  # apply floor function to column
df['rint_col'] = df['col1'].apply(np.rint)  # apply round to nearest integer function to column
df['trunc_col'] = df['col2'].apply(np.trunc)  # apply truncation function to column
df['fix_col'] = df['col3'].apply(np.fix)  # apply truncation towards zero function to column
df['isnan_col'] = df['col1'].apply(np.isnan)  # apply is NaN function to column
df['isfinite_col'] = df['col2'].apply(np.isfinite)  # apply is finite function to column
df['isinf_col'] = df['col3'].apply(np.isinf)  # apply is infinite function to column
df['logical_not_col'] = df['col1'].apply(np.logical_not)  # apply logical NOT function to column
df['cosh_col'] = df['col2'].apply(np.cosh)  # apply hyperbolic cosine function to column
df['sinh_col'] = df['col3'].apply(np.sinh)  # apply hyperbolic sine function to column
df['tanh_col'] = df['col1'].apply(np.tanh)  # apply hyperbolic tangent function to column
df['arccosh_col'] = df['col2'].apply(np.arccosh)  # apply inverse hyperbolic cosine function to column
df['arcsinh_col'] = df['col3'].apply(np.arcsinh)  # apply inverse hyperbolic sine function to column
df['arctanh_col'] = df['col1'].apply(np.arctanh)  # apply inverse hyperbolic tangent function to column
df['degrees_col'] = df['col2'].apply(np.degrees)  # convert radians to degrees for column
df['radians_col'] = df['col3'].apply(np.radians)  # convert degrees to radians for column
df['angle_col'] = df['col1'].apply(np.angle)  # calculate angle of complex number in column
df['conj_col'] = df['col2'].apply(np.conj)  # calculate complex conjugate of number in column
df['real_col'] = df['col3'].apply(np.real)  # get real part of complex number in column
df['add_col'] = np.add(df['col1'], df['col2'])  # element-wise addition of two columns
df['subtract_col'] = np.subtract(df['col2'], df['col3'])  # element-wise subtraction of two columns
df['multiply_col'] = np.multiply(df['col3'], 2)  # element-wise multiplication of column by scalar
df['divide_col'] = np.divide(df['col1'], df['col2'])  # element-wise division of two columns
df['mod_col'] = np.mod(df['col2'], 3)  # element-wise modulo operation on column by scalar
df['power_col'] = np.power(df['col3'], 2)  # element-wise exponentiation of column by scalar
df['sqrt_col'] = np.sqrt(df['col1'])  # element-wise square root of column
df['cbrt_col'] = np.cbrt(df['col2'])  # element-wise cube root of column
df['square_col'] = np.square(df['col3'])  # element-wise square of column
df['absolute_col'] = np.absolute(df['col1'])  # element-wise absolute value of column
df['reciprocal_col'] = np.reciprocal(df['col2'])  # element-wise reciprocal of column
df['exp_col'] = np.exp(df['col3'])  # element-wise exponential function of column
df['log_col'] = np.log(df['col1'])  # element-wise natural logarithm of column
df['log2_col'] = np.log2(df['col2'])  # element-wise base-2 logarithm of column
df['log10_col'] = np.log10(df['col3'])  # element-wise base-10 logarithm of column
df['mean_col'] = np.mean(df['col1'])  # calculate mean of column
df['median_col'] = np.median(df['col2'])  # calculate median of column
df['std_col'] = np.std(df['col3'])  # calculate standard deviation of column
df['var_col'] = np.var(df['col1'])  # calculate variance of column
df['min_col'] = np.min(df['col2'])  # find minimum value of column
df['max_col'] = np.max(df['col3'])  # find maximum value of column
df['argmin_col'] = np.argmin(df['col1'])  # find index of minimum value of column
df['argmax_col'] = np.argmax(df['col2'])  # find index of maximum value of column
df['unique_col'] = np.unique(df['col3'])  # find unique values of column
df['isin_col'] = np.isin(df['col1'], [1, 3, 5])  # check if elements of column are in a list
df['sort_col'] = np.sort(df['col2'])  # sort column in ascending order
df['argsort_col'] = np.argsort(df['col3'])  # find indexes to sort column in ascending order
df['clip_col'] = np.clip(df['col1'], 2, 4)  # clip column values to a specified range
df['around_col'] = np.around(df['col2'], decimals=1)  # round column values to a specified number of decimals
df['cumsum_col'] = np.cumsum(df['col3'])  # calculate cumulative sum of column
df['cumprod_col'] = np.cumprod(df['col1'])  # calculate cumulative product of column
df['diff_col'] = np.diff(df['col2'])  # calculate the discrete difference of column
df['gradient_col'] = np.gradient(df['col3'])  # calculate the gradient of column
df['trapz_col'] = np.trapz(df['col1'], x=df['col2'])  # integrate column using the composite trapezoidal rule
df['interp_col'] = np.interp(df['col2'], [6, 7, 8, 9, 10], [0, 1, 2, 3, 4])  # interpolate column values based on two arrays
df['polyfit_col'] = np.polyfit(df['col3'], df['col1'], deg=1)  # fit a polynomial of degree n to column
df['roots_col'] = np.roots([1, -2, 1])  # find the roots of a polynomial in column
df['polyval_col'] = np.polyval([1, 2, 1], df['col2'])  # evaluate a polynomial at specific values in column
df['histogram_col'] = np.histogram(df['col3'], bins=5)  # calculate a histogram of column values
df['bincount_col'] = np.bincount(df['col1'])  # count occurrences of each value in column
df['digitize_col'] = np.digitize(df['col2'], [6, 7, 8, 9, 10])  # return the indices of the bins to which column values belong
df['binomial_col'] = np.random.binomial(n=10, p=0.5, size=5)  # generate random binomial numbers for column
df['exponential_col'] = np.random.exponential(scale=1, size=5)  # generate random exponential numbers for column
df['normal_col'] = np.random.normal(loc=0, scale=1, size=5)  # generate random normal numbers for column
df['poisson_col'] = np.random.poisson(lam=1, size=5)  # generate random Poisson numbers for column
df['uniform_col'] = np.random.uniform(low=0, high=1, size=5)  # generate random uniform numbers for column
df['shuffle_col'] = np.random.shuffle(df['col1'])  # shuffle column values
df['seed_col'] = np.random.seed(0)  # set the seed for reproducibility of random values
df['correlate_col'] = np.correlate(df['col1'], df['col2'], mode='full')
```
